```python
import pandas as pd
airbnb = pd.read_csv('airbnb.csv')
```

## Using the method info()

Using the method info() we can obtain the answers for the following questions:

__Question 1__ - How many observations does the dataset airbnb have? <br>
__Question 2__ - Do we have missing values in the dataset airbnb?<br>
__Question 3__ - How many variables do not have any valid observations (all observations are missing)?<br>
__Question 4__ - How many integer variables do we have in our dataset?<br>

```python
airbnb.info()
```

Answer to question... <br>
__1__ - 18723 Observations <br>
__2__ - Yes (Not all variables have 18273 non-null observations)<br>
__3__ - 4 (The variables with 0 non-null counting)<br>
__4__ - 5 variables (room_id, survey_id, host_id, reviews and accomodates)<br>

## Using the method value_counts()

Using the method value_counts() we can obtain the answer for the following questions: <br>
__10__ - How many houses can accommodate more than 15 people? <br>
__11__ - How many houses accommodates only one person?

```python
airbnb['accommodates'].value_counts()
```

Answer to question... <br>
__10__ - 21 houses (sum of houses with 16 and 17 in accomodates) <br>
__11__ - 367 houses <br>
<hr>

## Using the method groupby()

Using the method value_counts() we can obtain the answer for the following questions: <br>
__12__ - What is the mean price for houses with an overall_satisfation of 5.0 (1 decimal houses)? <br>
__13__ - What is the mean price of a house that hosts 16 people?

```python
airbnb.groupby('overall_satisfaction')['price'].mean()
```

```python
airbnb.groupby('accommodates')['price'].mean()
```

Answer to question... <br>
__12__ - 170.2 <br>
__13__ - 745.7 <br>
<hr>

## Using the method corr()

Using the method corr() we can obtain the answer for the following questions: <br>
__14__ - What is the variable that has the higher correlation (Pearson) with the target (price)? <br>
__15__ - What is the correlation (pearson) between reviews and overall_satisfaction (2 decimal houses)? <br>

```python
airbnb.corr(method = 'pearson')
```

Answer to question... <br>
__14__ - accomodates (0.5) <br>
__15__ - 0.32 <br>
